### Setting up

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

### Flipped result

```python
model_version = '4'
# Load the data
with open(f'results/results_flipped_{model_version}.json') as f:
    data = json.load(f)
```

```python
# plot the data
from tasks.manager import task_list
plt.figure(figsize=(10, 5))
for task in task_list:
    normal_mean = data[task]['normal']['Classification'][0]
    normal_std = data[task]['normal']['Classification'][1]
    flipped_mean = data[task]['flipped']['Classification'][0]
    flipped_std = data[task]['flipped']['Classification'][1]
    plt.errorbar(task, normal_mean, normal_std, fmt='o', color='blue')
    plt.errorbar(task, flipped_mean, flipped_std, fmt='x', color='red', markersize=10)
plt.ylim(-0.05, 1.05)
plt.legend(['normal', 'flipped'])
plt.title(f'Normal vs Flipped (Classification) {data["model"]}')
```

```python
plt.figure(figsize=(10, 5))
for task in task_list:
    normal_mean = data[task]['normal']['Articulation'][0]
    normal_std = data[task]['normal']['Articulation'][1]
    flipped_mean = data[task]['flipped']['Articulation'][0]
    flipped_std = data[task]['flipped']['Articulation'][1]
    plt.errorbar(task, normal_mean, normal_std, fmt='o', color='blue')
    plt.errorbar(task, flipped_mean, flipped_std, fmt='x', color='red', markersize=10)
    plt.ylim(-0.05, 1.05)
plt.legend(['normal', 'flipped'])
plt.title(f'Normal vs Flipped (Articulation) {data["model"]}')
```

```python
plt.figure(figsize=(10, 5))
for task in task_list:
    class_mean = data[task]['normal']['Classification'][0]
    class_std = data[task]['normal']['Classification'][1]
    artic_mean = data[task]['normal']['Articulation'][0]
    artic_std = data[task]['normal']['Articulation'][1]
    plt.errorbar(task, class_mean, class_std, fmt='o', color='blue')
    plt.errorbar(task, artic_mean, artic_std, fmt='x', color='red', markersize=10)
plt.ylim(-0.05, 1.05)
plt.legend(['Classification', 'Articulation'])
plt.title(f'Accuracies (normal run) {data["model"]}')
```

```python
plt.figure(figsize=(10, 5))
for task in task_list:
    class_mean = data[task]['flipped']['Classification'][0]
    class_std = data[task]['flipped']['Classification'][1]
    artic_mean = data[task]['flipped']['Articulation'][0]
    artic_std = data[task]['flipped']['Articulation'][1]
    plt.errorbar(task, class_mean, class_std, fmt='o', color='blue')
    plt.errorbar(task, artic_mean, artic_std, fmt='x', color='red', markersize=10)
    plt.ylim(-0.05, 1.05)
plt.legend(['Classification', 'Articulation'])
plt.title(f'Accuracies (flipped run) {data["model"]}')
```
