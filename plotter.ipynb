```python
import json
import matplotlib.pyplot as plt
import numpy as np
model_version = '3'
# Load the data
with open(f'results_flipped_{model_version}.json') as f:
    data = json.load(f)

data
```

```python
# plot the data
from tasks.manager import task_list
plt.figure(figsize=(10, 5))
for task in task_list:
    normal_mean = data[task]['normal']['Classification'][0]
    normal_std = data[task]['normal']['Classification'][1]
    flipped_mean = data[task]['flipped']['Classification'][0]
    flipped_std = data[task]['flipped']['Classification'][1]
    plt.errorbar(task, normal_mean, normal_std, fmt='o', color='blue')
    plt.errorbar(task, flipped_mean, flipped_std, fmt='o', color='red')
plt.legend(['normal', 'flipped'])
plt.title(f'Classification Accuracy {data["model"]}')
```

```python
plt.figure(figsize=(10, 5))
for task in task_list:
    normal_mean = data[task]['normal']['Articulation'][0]
    normal_std = data[task]['normal']['Articulation'][1]
    flipped_mean = data[task]['flipped']['Articulation'][0]
    flipped_std = data[task]['flipped']['Articulation'][1]
    plt.errorbar(task, normal_mean, normal_std, fmt='o', color='blue')
    plt.errorbar(task, flipped_mean, flipped_std, fmt='o', color='red')
plt.legend(['normal', 'flipped'])
plt.title(f'Articulation Accuracy {data["model"]}')
```
