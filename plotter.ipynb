### Setting up

```python
import json
import matplotlib.pyplot as plt
import numpy as np
model_version = '4'
```

### Flipped result

```python
# Load the data
with open(f'results/results_flipped_{model_version}.json') as f:
    data = json.load(f)
```

```python
PRIMARY, SECONDARY, TERTIARY, PRIMARY_LIGHT, SECONDARY_LIGHT, TERTIARY_LIGHT = ((0.2823529411764706, 0.47058823529411764, 0.8156862745098039),
 (0.9333333333333333, 0.5215686274509804, 0.2901960784313726),
 (0.41568627450980394, 0.8, 0.39215686274509803),
 (0.6313725490196078, 0.788235294117647, 0.9568627450980393),
 (1.0, 0.7058823529411765, 0.5098039215686274),
 (0.5529411764705883, 0.8980392156862745, 0.6313725490196078))
```

```python
plt.figure(figsize=(15, 5))
for task in task_list:
    m1 = data[task]['flipped']['Classification'][0]
    s1 = data[task]['flipped']['Classification'][1]
    m2 = data[task]['flipped']['Articulation'][0]
    s2 = data[task]['flipped']['Articulation'][1]

    m3 = data[task]['normal']['Classification'][0]
    s3 = data[task]['normal']['Classification'][1]
    m4 = data[task]['normal']['Articulation'][0]
    s4 = data[task]['normal']['Articulation'][1]

    plt.errorbar(task, m3, s3, fmt='o', color=PRIMARY)
    plt.errorbar(task, m4, s4, fmt='x', color=SECONDARY, markersize=10, linestyle='-.')

    plt.errorbar(task+"\n(flipped)", m1, s1, fmt='o', color=PRIMARY)
    plt.errorbar(task+"\n(flipped)", m2, s2, fmt='x', color=SECONDARY, markersize=10, linestyle='-.')

    plt.ylim(-0.05, 1.05)
plt.legend(['Classification', 'Articulation'],bbox_to_anchor=(1.0, 1.0), loc='upper left', )
#make legend as wide as the figure

plt.title(f'Accuracies (flipped run) {data["model"]}')
```
